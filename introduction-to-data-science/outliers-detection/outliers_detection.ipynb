## Paired density and scatterplot matrix

Vamos verificar todas as correlações possíveis entre as variáveis numéricas considerando o gráfico Scatterplot Matrix

```python
g = sns.PairGrid(df._get_numeric_data(), diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot, lw=3)
```

Parece que as variáveis mais correlacionadas são as notas referentes aos 3 períodos (G1, G2 e G3). Vamos olhar mais de perto e imprimir o coeficiente de correlação de Pearson dessas variáveis.

```python
plt.figure(figsize=(10,7))
sns.regplot(x="G1", y="G2", data=df)
print("Pearson correlation coefficient: " + str(stats.pearsonr(df['G1'], df['G2'])[0]))
```

```python
plt.figure(figsize=(10,7))
sns.regplot(x="G2", y="G3", data=df)
print("Pearson correlation coefficient: " + str(stats.pearsonr(df['G2'], df['G3'])[0]))
```

```python
plt.figure(figsize=(10,7))
sns.regplot(x="G1", y="G3", data=df)
print("Pearson correlation coefficient: " + str(stats.pearsonr(df['G1'], df['G3'])[0]))
```

Conforme vimos anteriormente, há correlação forte entre as notas dos 3 períodos considerados (acima de 0.82). A correlação maior foi entre o segundo e terceiro período (0.91). Provavelmente, para esse dataset, o comportamento de estudos dos alunos não estão sofrendo grandes alterações ao longo dos períodos e isso se reflete na nota.

```python
plt.figure(figsize=(10,7))
g = sns.lmplot(x="G2", y="G3", hue="school", data=df, legend_out = True)
new_labels = ['Gabriel Pereira', 'Mousinho da Silveira']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)
```

Vemos que não há muita diferença entre as correlações das duas escolas do dataset.

```python
plt.figure(figsize=(10,7))
sns.regplot(x="Fedu", y="Medu", data=df)
print("Pearson correlation coefficient: " + str(stats.pearsonr(df['Fedu'], df['Medu'])[0]))
```

Há uma correlação média entre o nível de estudo da mãe comparado com o nível de estudo do pai.

## Boxplot (idade)

Vamos analisar o perfil dos alunos em função das idades.

```python
plt.figure(figsize=(10,7))
sns.boxplot(y='age' , x='freetime', data=df, palette="Paired")
plt.xticks(np.arange(5), ("1 - very low", "2", "3", "4", "5 - very high"))
plt.xlabel('Free time after school')
plt.ylabel('Age of Users')
plt.title('Free time after school vs. age')
sns.despine()
```

Notamos alguns outliers para os alunos que dizem possuir bastante tempo livre após a escola. E esse comportamento se dá para alunos mais velhos.

```python
plt.figure(figsize=(10,7))
sns.boxplot(y='age' , x='studytime', data=df, palette="Paired")
plt.xticks(np.arange(4), ("<2 hours", "2 to 5 hours", "5 to 10 hours", "4 - >10 hours"))
plt.xlabel('Weekly study time')
plt.ylabel('Age of Users')
plt.title('Weekly study time vs. age')
sns.despine()
```

- Notamos alguns outliers para os alunos mais velhos que dizem estudar menos de 2 horas.
- A mediana dos alunos que estudam mais de 10 horas semanais é de 16 anos. Portanto, ao levar em conta a idade mediana, parece que esses alunos estão estudando mais do que os alunos com 17 anos.

## Heatmap

```python
df2 = df.drop_duplicates(['freetime','studytime'])
df2 = df2.pivot("freetime", "studytime", "G3")
```

```python
plt.figure(figsize=(10,7))
ax = sns.heatmap(df2, annot=True, fmt="d", linewidths=.7, cmap="hot")
```

O aluno que diz ter bastante tempo livre e que simultaneamente estuda pouco possui a pior nota (7).

```python
df3 = df.drop_duplicates(['freetime','goout'])
df3 = df3.pivot("freetime", "goout", "G3")
```

```python
plt.figure(figsize=(10,7))
ax = sns.heatmap(df3, annot=True, fmt="d", linewidths=.20, cmap="hot")
```

Vemos que os extremos são ruins nesse caso. Ou seja, alunos que tem pouco tempo livre e não saem com os amigos tem a pior nota. Assim como alunos que saem bastante com os amigos e possuem bastante tempo livre.

```python
plt.figure(figsize=(10,7))
sns.regplot(x="failures", y="G3", data=df)
print("Pearson correlation coefficient: " + str(stats.pearsonr(df['Fedu'], df['Medu'])[0]))
```
